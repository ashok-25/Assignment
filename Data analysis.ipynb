```python
# Converting budget and gross values from dollers to million dollers

data['budget']=data['budget']/1000000
data['Gross']=data['Gross']/1000000
```

# Subtask-2

```python
# Making profit column 

profit_df=data[['Title','budget','Gross']]
profit_df['Profit']=profit_df['Gross']-profit_df['budget']  
profit_df
```

```python
# sorting dataframe in descending order
profit_df=profit_df.sort_values('Profit',ascending=False,ignore_index=True)
profit_df.head()
```

```python
# extract top10 profit making movies

top10=profit_df[0:10]
print(top10)
```

We can see that star wars making highest profit of 691.66 million dollers

```python
#plotting scatter plot
sns.scatterplot(x='budget',y='Profit',data=top10)
```

From above plot we can see that there is no linear relation between budget and profit.

We can see we can get high profits with high budget.

We can get profits with less budget also.

```python
# extract top10 profit making movies

neg_prof=profit_df[-10:-1].sort_values('Profit',ascending=True,ignore_index=True)
print(neg_prof)
```

We can see that The Little Prince making highest loss of 79.86 million dollers

```python
df=data[['genre_1','genre_2','genre_3','CVotes10', 'CVotes09', 'CVotes08', 'CVotes07', 'CVotes06', 'CVotes05',
       'CVotes04', 'CVotes03', 'CVotes02', 'CVotes01', 'CVotesMale',
       'CVotesFemale', 'CVotesU18', 'CVotesU18M', 'CVotesU18F', 'CVotes1829',
       'CVotes1829M', 'CVotes1829F', 'CVotes3044', 'CVotes3044M',
       'CVotes3044F', 'CVotes45A', 'CVotes45AM', 'CVotes45AF', 'CVotes1000',
       'CVotesUS', 'CVotesnUS', 'VotesM', 'VotesF', 'VotesU18', 'VotesU18M',
       'VotesU18F', 'Votes1829', 'Votes1829M', 'Votes1829F', 'Votes3044',
       'Votes3044M', 'Votes3044F', 'Votes45A', 'Votes45AM', 'Votes45AF',
       'Votes1000', 'VotesUS', 'VotesnUS']]
df
```

```python
# Adding cnt column to the dataframe by genre
df['cnt']=1
```

```python
df
```

```python
df_g1=df.groupby('genre_1').sum()
df_g1
```

```python
df_g2=df.groupby('genre_2').sum()
df_g2
```

```python
df_g3=df.groupby('genre_3').sum()
df_g3
```

```python
# Adding three dataframes

df_add_1=df_g1.add(df_g2,fill_value=0)
df_add_1
```

```python
df_add=df_add_1.add(df_g3,fill_value=0)
df_add
```

```python
# mean of the each voters column
for feature in df_add.columns:
    df_add[feature]=(df_add[feature]/df_add['cnt']).astype('int32')
   
```
